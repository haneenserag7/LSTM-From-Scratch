```python
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

X = [1, 2, 3]  

input_dim = 1
hidden_dim = 1

np.random.seed(1)

Wf = np.random.randn(hidden_dim, input_dim)
Wi = np.random.randn(hidden_dim, input_dim)
Wo = np.random.randn(hidden_dim, input_dim)
Wc = np.random.randn(hidden_dim, input_dim)

f = np.random.randn(hidden_dim, hidden_dim)
i = np.random.randn(hidden_dim, hidden_dim)
o = np.random.randn(hidden_dim, hidden_dim)
c = np.random.randn(hidden_dim, hidden_dim)

bf = np.zeros((hidden_dim, 1))
bi = np.zeros((hidden_dim, 1))
bo = np.zeros((hidden_dim, 1))
bc = np.zeros((hidden_dim, 1))

h = np.zeros((hidden_dim, 1))
c = np.zeros((hidden_dim, 1))


print("LSTM Forward Pass ")

for t, x in enumerate(X):

    x = np.array([[x]])

    f = sigmoid(Wf @ x + f @ h + bf)

    i = sigmoid(Wi @ x + i @ h + bi)

    c_tilde = tanh(Wc @ x + c @ h + bc)

    c = f * c + i * c_tilde

    o = sigmoid(Wo @ x + o @ h + bo)

    h = o * tanh(c)

    print(f"\nTime Step {t+1}")
    print("Forget gate:", f)
    print("Input gate:", i)
    print("Candidate state:", c_tilde)
    print("Cell state:", c)
    print("Output gate:", o)
    print("Hidden state:", h)
```

```python
Wy = np.array([[1.2]])
by = np.array([[0.5]])

y_hat = Wy @ h + by

print("Final Prediction")
print("Predicted next value:", float(y_hat))
```
